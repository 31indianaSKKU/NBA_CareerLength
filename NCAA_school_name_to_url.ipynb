```python
import requests
from bs4 import BeautifulSoup

# 웹 페이지에 접속하여 HTML 데이터 가져오기
url = "https://www.sports-reference.com/cbb/schools/#all_NCAAM_schools"
response = requests.get(url)
html_data = response.text

# BeautifulSoup을 사용하여 HTML 파싱
soup = BeautifulSoup(html_data, "html.parser")

# 학교 이름 변경을 위한 딕셔너리 초기화
schools_dict = {}

# 학교 테이블에서 데이터 가져와서 딕셔너리에 추가
table = soup.find("table", id="NCAAM_schools")
rows = table.tbody.find_all("tr")
for row in rows:
    cells = row.find_all("td")
    if len(cells) > 0:
        original_school = cells[0].text.strip()
        
        # schools 값 변경 방법1: 괄호와 점 제거
        modified_school = original_school.replace("(", "").replace(")", "").replace(".", "").replace("&","")
        
        # schools 값 변경 방법2: 공백을 "-"로 대체
        modified_school = modified_school.replace(" ", "-")
        
        # 대문자를 소문자로 변경
        modified_school = modified_school.lower()
        
        schools_dict[original_school] = modified_school

# 결과 출력
for original_school, modified_school in schools_dict.items():
    print(f"{original_school} -> {modified_school}")
```

```python
import requests
from bs4 import BeautifulSoup

# 학교 이름 변경을 위한 딕셔너리 초기화
schools_dict = {}

# Step 1: https://www.sports-reference.com/cbb/seasons/men/1991-ratings.html 페이지에서 학교 이름 가져와서 딕셔너리에 추가
url = "https://www.sports-reference.com/cbb/seasons/men/1991-ratings.html"
response = requests.get(url)
html_data = response.text
soup = BeautifulSoup(html_data, "html.parser")

table = soup.find("table", id="ratings")
rows = table.tbody.find_all("tr")
for row in rows:
    cells = row.find_all("td")
    if len(cells) > 0:
        original_school = cells[0].text.strip()
        
        # school 값 변경 방법1: 괄호와 점 제거
        modified_school = original_school.replace("(", "").replace(")", "").replace(".", "")
        
        # school 값 변경 방법2: 공백을 "-"로 대체
        modified_school = modified_school.replace(" ", "-")
        
        # school 값 변경 방법3: 소문자로 변경
        modified_school = modified_school.lower()
        
        schools_dict[original_school] = modified_school

# Step 2: 1991부터 2021까지의 연도에 대해 반복 실행하여 딕셔너리에 추가
for yrs in range(1992, 2024):
    url = f"https://www.sports-reference.com/cbb/seasons/men/{yrs}-ratings.html"
    response = requests.get(url)
    if response.status_code == 200:
        if yrs not in schools_dict:
            schools_dict[yrs] = url
            
# 알파벳 순으로 딕셔너리 정렬
sorted_dict = dict(sorted(schools_dict.items(), key=lambda x: str(x[0])))

# 정렬된 딕셔너리 출력
for key, value in sorted_dict.items():
    print(f"{key}: {value}")

# # 결과 출력
# for key, value in schools_dict.items():
#     print(f"{key} -> {value}")
```

```python
import requests
from bs4 import BeautifulSoup

# Step 1: School value modifications
url = "https://www.sports-reference.com/cbb/seasons/men/1991-ratings.html"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
table = soup.find("table", id="ratings")
schools_dict = {}

for row in table.find_all("tr"):
    cells = row.find_all("td")
    if len(cells) > 0:
        school = cells[1].text.strip()
        modified_school = school.replace("(", "").replace(")", "").replace(".", "").replace(" ", "-").lower()
        schools_dict[school] = modified_school

# Step 2: Iterate over years and check if the school is in the dictionary
for yrs in range(1991, 2022):
    url = f"https://www.sports-reference.com/cbb/seasons/men/{yrs}-ratings.html"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    table = soup.find("table", id="ratings")

    for row in table.find_all("tr"):
        cells = row.find_all("td")
        if len(cells) > 0:
            school = cells[1].text.strip()
            if school not in schools_dict:
                modified_school = school.replace("(", "").replace(")", "").replace(".", "").replace(" ", "-").lower()
                schools_dict[school] = modified_school

# Print the number of schools in the dictionary
print(f"Number of schools: {len(schools_dict)}")
```

```python
#학교이름과 url내의 학교이름 mapping!!
#기간설정
import requests
from bs4 import BeautifulSoup

# Define the range of years
start_year = 2019
end_year = 2023

# Create a dictionary to store the school names and URLs
school_mapping = {}

for year in range(start_year, end_year + 1):
    url = f"https://www.sports-reference.com/cbb/seasons/men/{year}-ratings.html"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    table = soup.find("table", id="ratings")

    for row in table.find_all("tr"):
        cells = row.find_all("td")
        if len(cells) > 0:
            school_name = cells[0].text.strip()
            link = cells[0].find("a")["href"]
            link_parts = link.split("/")
            desired_string = link_parts[3]

            # Check if the school name already exists in the dictionary
            if school_name not in school_mapping:
                school_mapping[school_name] = desired_string

# Print the school mapping
for school_name, url_school in school_mapping.items():
    print("School Name:", school_name)
    print("URL School:", url_school)
    print()
```
